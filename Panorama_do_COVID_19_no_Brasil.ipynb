# Panorama do COVID-19 no Brasil

O mundo foi assombrado por uma pandemia de um coronavírus descoberto no final de 2019 pela primeira vez na província de Wuhan na China. Mas que explodiu e parou o mundo entre Fevereiro e Março de 2020.

As variantes iniciais eram transmitidas principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, as novas variantes como a delta por conterem uma carga viral maior e adaptações do próprio vírus aumentou o nível de transmissão se tornando ainda mais perigoso, e talvez até impossibilitando a possibilidade de acabar o vírus através de imunidade de rebanho com a vacinação. Porém a mesma continua sendo fundamental por proteger o indíviduo e os que cercam.

Estudos continuam sendo realizados no mundo todo, entre eles os que levaram ao desenvolvimento das mais diversas vacinas

Agora é possível colocar a ordem dos países com mais casos totais:

Colocando as mortes por país para verificarmos os países com mais mortes:

Nos gráficos abaixos ordenamos os países em que ocorreram mais mortes absolutas e em seguida por mais mortes por milhão: 

```
# os cinco países que mais tiveram mortes
five = df.loc[df.date=='2021-08-25',['location','total_deaths']].sort_values(ascending=False, by = 'total_deaths')[0:5]

#visualizaçao
fig,ax = plt.subplots(figsize=(10,6))
ax.bar('location','total_deaths',data= five ,color = 'red')
ax.set_title('Líderes em mortes absolutas por covid 19')
ax.set_ylabel('Mortes')
fig.show()

fig.savefig('Mais mortais')

# os cinco países que mais tiveram mortes por milhão
five = df.loc[df.date=='2021-08-25',['location','total_deaths_per_million']].sort_values(ascending=False, by = 'total_deaths_per_million')[0:10]

#visualizaçao
fig,ax = plt.subplots(figsize=(18,6))
ax.bar('location','total_deaths_per_million',data= five ,color = 'red')
ax.set_title('Líderes em mortes por milhão por covid 19')
ax.set_ylabel('Mortes por milhão')
fig.show()

fig.savefig('Mais mortais')
```

Para descobrir o primeiro dia de vacinação só precisamos usar mais uma vez a função loc que quando o número for maior que zero quer dizer que a primeira vacina foi aplicada, visto que não é possível para tal variável ser negativa:

```
df_brasil.loc[df.people_vaccinated_per_hundred > 0]
```

```
# Plotar gráficos de linha em escalas linear e logarítmica para os casos totais e mortes totais.
x = list(range(566))
z = df_brasil['total_cases']
z2 = df_brasil['total_deaths']

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 9))
ax1.set_title('Casos totais no Brasil')
ax1.set_xlabel('Dias')
ax1.set_ylabel('Casos')
ax1.plot(x, z, color='red')

ax2.set_title('Mortes totais no Brasil')
ax2.set_xlabel('Dias')
ax2.set_ylabel('Mortes')
ax2.plot(x, z2, color='red')

ax3.set_title('Casos totais no Brasil (Em log)')
ax3.set_xlabel('Dias')
ax3.set_ylabel('Casos')
ax3.set_yscale('log')
ax3.plot(x, z, color='red')

ax4.set_title('Mortes totais no Brasil (Em log')
ax4.set_xlabel('Dias')
ax4.set_ylabel('Mortes')
ax4.set_yscale('log')
ax4.plot(x, z2, color='red')
plt.tight_layout()
```

### Avanço da Vacinação no Brasil

Plotaremos gráficos semelhantes aos anteriores com a média móvel de mortes a vacinação por 100 habitantes, para analisar se há indicativos de correlação entre os dados.

```
x = list(range(566))
z = df_brasil['new_deaths_smoothed']
z2 = df_brasil['people_vaccinated_per_hundred']

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 9))
ax1.set_title('Média móvel de mortes')
ax1.set_xlabel('Dias')
ax1.set_ylabel('Média móvel')
ax1.plot(x, z, color='red')

ax2.set_title('Vacinação por 100 habitantes')
ax2.set_xlabel('Dias')
ax2.set_ylabel('Vacinados por 100 habitantes')
ax2.plot(x, z2, color='green')

ax3.set_title('Média móvel de mortes (Em log)')
ax3.set_xlabel('Dias')
ax3.set_ylabel('Média móvel')
ax3.set_yscale('log')
ax3.plot(x, z, color='red')

ax4.set_title('Vacinação por 100 habitantes (Em log')
ax4.set_xlabel('Dias')
ax4.set_ylabel('Vacinados por 100 habitantes')
ax4.set_yscale('log')
ax4.plot(x, z2, color='green')
plt.tight_layout()
```

Aparentemente conforma a vacinação aumentou pelo dia 400 a média móvel de mortes começou a decair.
